```python
import json

# JSON 파일에서 도시 데이터를 로드합니다.
json_file_path = "./files/city.list.json"
with open(json_file_path, encoding='utf-8') as json_file:
    city_data = json.load(json_file)

# 한국에 있는 도시만 선택할 리스트를 초기화합니다.
korean_cities = []

# 도시 데이터를 순회하면서 국가 코드가 "KR"인 도시를 선택합니다.
for city_info in city_data:
    country_code = city_info.get('country')  # 국가 코드 가져오기
    if country_code == "KR":
        korean_cities.append(city_info['name'])

# 선택된 한국 도시 목록을 출력합니다.
for city_info in korean_cities:
    print(city_info)
    
# JSON 파일로 저장
with open("./files/korean_cities.json", "w", encoding="utf-8") as json_file:
    json.dump(korean_cities, json_file, ensure_ascii=False, indent=4)
```
