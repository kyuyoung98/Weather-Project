# 동아시아(경도 100~160, 위도 0 ~ 60) 도시 이름 생성

```python
import json

# JSON 파일에서 도시 데이터를 로드합니다.
json_file_path = "./files/city.list.json"
with open(json_file_path, encoding='utf-8') as json_file:
    city_data = json.load(json_file)

# 원하는 경도와 위도 범위를 정의합니다.
# 예: 동아시아 지역의 경도 범위 (100 도 ~ 160 도), 위도 범위 (0 도 ~ 60 도)
min_longitude = 100
max_longitude = 160
min_latitude = 0
max_latitude = 60

# 선택한 도시를 저장할 리스트를 초기화합니다.
selected_cities = []

# 도시 데이터를 순회하면서 경도와 위도 범위 내에 있는 도시를 선택합니다.
for city_info in city_data:
    lon = city_info['coord']['lon']
    lat = city_info['coord']['lat']
    
    if min_longitude <= lon <= max_longitude and min_latitude <= lat <= max_latitude:
        selected_cities.append(city_info['name'])

# 선택된 도시 목록을 출력합니다.
for city_name in selected_cities:
    print(city_name)
```

```python
# JSON 파일로 저장
with open("./files/selected_cities.json", "w", encoding="utf-8") as json_file:
    json.dump(selected_cities, json_file, ensure_ascii=False, indent=4)
```
