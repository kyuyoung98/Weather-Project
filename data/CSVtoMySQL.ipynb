```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
df = pd.read_csv("./files/korea_weather/20230920_Weather.csv")
```

```python
df.head()
```

```python
# 칼럼명 변환
df = df.rename(columns={
    'City ID': 'weather_id',
    'City Name': 'city_name',
    'Lon': 'lon',
    'Lat': 'lat',
    'Weather Main': 'weather_main',
    'Temperature (K)': 'temperature',
    'Humidity (%)': 'humidity',
    'Wind Speed (m/s)': 'wind_speed',
    'Clouds (%)': 'clouds',
    'Date (Unix)': 'date'
})
```

```python
# 필요한 열만 선택
selected_columns = ['weather_id', 'date', 'city_name', 'lon', 'lat', 'weather_main', 'temperature', 'humidity', 'wind_speed', 'clouds']
df = df[selected_columns]
```

```python
df.head()
```

```python
# 'date' 열을 날짜 형식으로 파싱
df['date'] = pd.to_datetime(df['date'])

# 'date' 열을 'YYYY-MM-DD' 형식으로 변환
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
```

```python
df.head()
```

```python
# DB 설정
from sqlalchemy import create_engine

db_url = 'mysql+pymysql://root:rootroot@localhost:3306/weather' 
engine = create_engine(db_url)
```

```python
df.to_sql('weather', engine, if_exists='replace', index=False)
```
